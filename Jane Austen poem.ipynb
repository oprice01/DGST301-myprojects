```python
with open('JaneAusten.txt', 'r') as file:
    text = file.read()
    import random
    from textblob import TextBlob
blob = TextBlob(text)
adjectives = []
nouns = []
verbs = []
pronouns = []
adverbs = []
prepositions = []
modals = []
for word,pos in blob.tags:
    print(word,pos)
    if (pos == 'JJ'):
        adjectives.append(word)
    if (pos == 'NN'):
        nouns.append(word)
    if (pos == 'VBD'):
        verbs.append(word)
    if (pos == 'PRP'):
        pronouns.append(word)
    if (pos == 'RB'):
        adverbs.append(word)
    if (pos == 'IN'):
        prepositions.append(word)
    if (pos == 'MD'):
        modals.append(word)
```

```python
for i in range(8):
    p1 = random.choice(pronouns)
    a1 = random.choice(adjectives)
    n1 = random.choice(nouns)
    n2 = random.choice(nouns)
    v1 = random.choice(verbs)
    ad1 = random.choice(adverbs)
    pr1 = random.choice(prepositions)
    print(p1 + " " + ad1 + " " + v1 + " " +  n1  + " " + pr1 + " " + a1  + " " + n2)
```
